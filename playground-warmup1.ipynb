```python
import datasets
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from torchvision import transforms
```

```python
custom_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),
])

practice_png_dataset = datasets.PracticePNGJPG("data/Practice_PNGandJPG/chestimage_PNG/", transform=custom_transform)
practice_png_dataloader = DataLoader(practice_png_dataset, batch_size=4, shuffle=True)
```

```python
for png_batch_tensor in practice_png_dataloader:
    batch_array = png_batch_tensor.permute(0, 2, 3, 1).numpy()
    print(f"Feature batch shape: {batch_array.shape[0]}")
    fig, axs = plt.subplots(1, batch_array.shape[0], figsize=(3*batch_array.shape[0],3))
    for i in range(batch_array.shape[0]):
        axs[i].imshow(batch_array[i], cmap=plt.cm.bone)
        axs[i].axis('off')
    plt.show()
```
