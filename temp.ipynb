```python
import matplotlib.pyplot as plt
import numpy as np

# Define the values
auc = [0.9026, 0.6376, 0.8841, 0.9336, 0.8937, 0.9188, 0.8578, 0.8161, 0.8713, 0.9658,
       0.9528, 0.9002, 0.6553, 0.9601]
diseases = ['No Finding', 'Enlarged Cardiomediastinum', 'Cardiomegaly', 'Lung Opacity',
            'Lung Lesion', 'Edema', 'Consolidation', 'Pneumonia', 'Atelectasis', 'Pneumothorax',
            'Pleural Effusion', 'Pleural Other', 'Fracture', 'Support Devices']

# Sort by AUC values
sorted_data = sorted(zip(auc, diseases), reverse=True)

# Unpack sorted data
auc_sorted, diseases_sorted = zip(*sorted_data)

# Calculate average AUC
average_auc = np.mean(auc_sorted)

# Create bar chart
plt.figure(figsize=(12, 8))
bars = plt.bar(diseases_sorted, auc_sorted, color='skyblue')

# Adding values on top of bars
for i, value in enumerate(auc_sorted):
    plt.text(i, value + 0.005, f'{value:.4f}', ha='center')

# Plotting average line
plt.axhline(y=average_auc, color='orange', linestyle='--', label=f'Average AUC: {average_auc:.4f}')

# Plotting threshold line
plt.axhline(y=0.9, color='red', linestyle='--', label='Threshold (0.9)')

# Adding title and labels
plt.title('AUC Scores for Different Diseases (Sorted)')
plt.xlabel('Diseases')
plt.ylabel('AUC Score')
plt.xticks(rotation=45, ha='right')

# Adding legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
